```python
import pandas as pd
import numpy as np
```

```python
df_siswa = pd.read_csv('https://raw.githubusercontent.com/vanindraa/vanindrarepo/Dataset/data_siswa.csv.csv')

df_siswa.head()
```

```python
df_siswa.info()
```

```python
df_siswa['Tinggi Badan (m)'] = df_siswa['Tinggi Badan (cm)']/100

df_siswa.head()
```

1. Hitunglah BMI setiap siswa dengan rumus BMI = Berat Badan(kg)/(Tinggi Badan(m)**2)

```python
BMI = df_siswa.assign(BMI = df_siswa['Berat badan (kg)']/(df_siswa['Tinggi Badan (m)']**2))

BMI.sample(10)
```

2. Update dataframe df_siswa dengan
menambahkan hasil perhitungan soal nomor
1, beri nama kolom BMI.

```python
df_siswa['BMI'] = df_siswa['Berat badan (kg)']/(df_siswa['Tinggi Badan (m)']**2)
```

```python
df_siswa.info()
```

```python
df_siswa.sample(5)
```

##### 3. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.
#### a. Normal: 18.5-24.9
#### b. Overweight: 25.0-29.9
#### c. Obesity: 30.0-34.9
#### d. Underweight: < 18.5

```python
df_siswa['Status_BMI'] = np.where(df_siswa['BMI'] <=34.9,'Obesity','Obesity')
df_siswa['Status_BMI'] = np.where(df_siswa['BMI'] <=29.9,'Overweight',df_siswa['Status_BMI'])
df_siswa['Status_BMI'] = np.where(df_siswa['BMI'] <=24.9,'Normal',df_siswa['Status_BMI'])
df_siswa['Status_BMI'] = np.where(df_siswa['BMI'] <18.5,'Underweight',df_siswa['Status_BMI'])
```

```python
df_siswa.sample(10)
```

```python
df_siswa.groupby('Status_BMI').count()[['Nama']]
```

#### 5. Filter dataframe df_siswa berdasarkan:
#### a. Status BMI di luar Normal.
#### b. Tinggi Badan > 169 cm dan Berat Badan > 60 kg
#### c. Nama yang mengandung huruf “y”.
#### d. Siapa saja yang berat badannya kurang dari 60 kg?

A -> Filtered out those with 'Normal' BMI

```python
status_filter = df_siswa.loc[df_siswa['Status_BMI']!='Normal']
status_filter
```

B -> Tinggi badan diatas 169cm dan berat badan diatas 60kg

```python
height_filter = df_siswa.loc[(df_siswa['Tinggi Badan (cm)'] > 169) & (df_siswa['Berat badan (kg)'] > 60)]

height_filter
```

C -> Nama yang mengandung huruf 'y'

```python
y_name = df_siswa[df_siswa.Nama.str.contains('y')]
y_name
```

D -> Siswa dengan berat badan di bawah 60kg

```python
weight_filter = df_siswa.loc[df_siswa['Berat badan (kg)'] < 60]

weight_filter
```
